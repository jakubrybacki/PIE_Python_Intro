# Zadanie - zmiana częstotliwości
Zadanie pokaże, jak w pandas można agregowanć danych do większych częstotliwości. Do naszej dyspozycji są informacje dotyczące dziennych zmian cen na rynkach hurtowych wszystkich państw UE.

Pochodzą one ze strony [Ember Climate](https://ember-climate.org/data-catalogue/european-wholesale-electricity-price-data/)

Będziemy ściągać dane na podstawie kopii z repozytorium Github:


```python
URL  = "https://raw.githubusercontent.com/jakubrybacki/Karowa_Python_Introduction/main/Zbiory%20cwiczen%20pandas/Zadanie%201%20-%20agregowanie%20danych/european_wholesale_electricity_price_data_daily-4.csv"
```

## Polecenie 1:
Zaimportuj bibliotekę pandas. Wykorzystaj adres URL u góry do wczytania informacji do ramki danych. Wyświetl sturkturę pliku.

```python
import pandas as pd

dane = pd.read_csv(URL)
dane.info()
```

## Polecenie 2:
Kolumna date przedstawiana jest jako objekt. Chcemy, aby Python rozpoznawał, że to data. Wykorzystajmy funkcje *to_datetime()* - [dokumentacja](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html)

```python
dane.head(5)
dane.Date = pd.to_datetime(dane.Date, format='%Y-%m-%d', errors='ignore')
dane.info()
```

## Polecenie 3:
Dane mają format tabeli dobry do wykorzystania tabeli przestawnej - kolejne kraje mają informacje przedstawiane w tzw. formie długiej. Chcemy przekształcić ją do formy szerokiej, gdzie:


*   W wierszach będziemy mieć daty.
*   W kolumnach nazwy krajów.
*   Jako wartości ceny elektryczności.



```python
dane_szerokie = dane.pivot(index='Date', columns='Country', values='Price (EUR/MWhe)')
dane_szerokie.head(5)
```

## Polecenie 4:
Wybierz z ramki dane dla Francji, Niemiec i Polski. Przedstaw je na wykresie liniowym.


```python
dane_wykres1 = dane_szerokie[["France", "Germany" ,"Poland"]]
dane_wykres1.plot.line()
```

## Polecenie 5:
Przekształć dane dzienne do miesięcznych - wykorzystaj komendę *resample*. Pomoże [dokumentacja](https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html).

Wyświetl pierwsze 5 rekordów.

```python
dane_miesieczne = dane_szerokie.resample("M").mean()
dane_miesieczne.head(5)
```

## Polecenie 6:
Stwórz analogiczny wykres jak w poleceniu 4 dla danych miesięcznych

```python
dane_wykres2 = dane_miesieczne[["France", "Germany" ,"Poland"]]
dane_wykres2.plot.line()
```
