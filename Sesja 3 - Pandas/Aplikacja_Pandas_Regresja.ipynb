### Polecenie 3:
Na podstawie tych danych stworzymy 2 ramki, które zawierać będa odpowiedzi na pojedyncze pytania w różnych krajach. Będą to kolejno:

1. Pierwsza - pytania o bieżącą aktywność firmy. Takie dane zawierają w nazwie sekwencję: 'TOT.1.BS.M'
2. Druga - pytania o wielkość zamówień. Tu sekwencja jest następująca: 'TOT.2.BS.M'

### Polecenie 4:
Poniżej znajdziecie funkcję, która:
1. Wyciąga z ramki wybraną serię danych po nazwie. Wykorzystamy ją dla ramki ***pytanie1***
2. Robi podobne działanie dla drugiej ramki - u nas będzie to ***pytanie2***
3. Tworzy model regresji liniowej gdzie zmienna z pierwszej ramki jest wyjaśniana przez zmienną z drugiej ramki.
4. Zwraca dwa parametry z równania: stałą i współczynnik przy zmiennej z ramki pytanie2 w postaci słownika.  

O wykorzystaniu regresji możecie poczytać tutaj:
https://realpython.com/linear-regression-in-python/

Oczywiście, ta wiedza nie będzie potrzebna nam do wykonania zadania - naszym celem będzie wykonanie już napisanej funkcji dla poszczególnych elementów listy.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

def pobierzParametry(zbior_wyjasniany, nazwa_serii_wyjasnianej,
                     zbior_objasniajacy, nazwa_serii_objasniającej):

  # Zmienia tablice pandas w obiekty numpy, które są wymagane przy pracach z modelami statystycznymi
  y = zbior_wyjasniany[nazwa_serii_wyjasnianej].to_numpy()
  x = zbior_objasniajacy[nazwa_serii_objasniającej].to_numpy()

  # Seria pandas to tablica jednowymiarowa - w regresji liniowej konieczne jest jej przekształcenie do 2D
  # Robią to te linijki kodu:
  y	= y.reshape(-1, 1)
  x = x.reshape(-1, 1)

  # Dopasowanie modelu
  model = LinearRegression().fit(x, y)

  # Parametry modelu trafiają do słownika
  slownik_parametry = {
      "stala": model.intercept_[0],
      "wspolczynnik": model.coef_[0,0]
  }
  return slownik_parametry

# Przykładowe wywołanie:
parametryPL = pobierzParametry(pytanie1, "INDU.PL.TOT.1.BS.M", pytanie2, "INDU.PL.TOT.2.BS.M")
print(parametryPL)
```

Wywołaj funkcję dla pięciu krajów zapisanych na liście.
*    Potrzebujemy dynamicznie tworzyć nazwy zmiennych, które przekazywać będziemy do funkcji
*   Jako wynik chcemy uzyskać krotkę w której pierwszym elementem byłby skrót krajów z listy *kraje_do_sprawdzenia*, a drugim elementem słownik parametrów uzyskany z funkcji.

Pytanie: W którym kraju współczynnik jest największy?

```python
kraje_do_sprawdzenia = ["PL", "CZ", "SK", "DE","HU"]

wynik = []
for i in kraje_do_sprawdzenia:
  wyjasniana = "INDU." + i + ".TOT.1.BS.M"
  wyjasniajaca = "INDU." + i + ".TOT.2.BS.M"
  parametry = pobierzParametry(pytanie1, wyjasniana, pytanie2, wyjasniajaca)

  wynik.append((i, parametry))

print(wynik)
```
