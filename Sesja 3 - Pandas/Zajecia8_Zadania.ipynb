### Zadanie 2
Napisz pętle które stworzą listę informacji o polskich laureatach:
* pokojowej nagrody Nobla.
* literackiej nagrody Nobla po 1980 roku

Dane o każdym laureacie przetrzymywane powinny być w krotce - tak jak w liście pol_Nobel.

Rozwiązanie podobnego problemu w Pandas przedstawiłem na dole.

```python
year =[1903,	1905,	1907,	1911,	1924,	1944,	1950,	1977,	1978,	1978, 1980,	1981,	1983,	1992,	1994,	1995,	1996,	2007, 2018]
who = ["Maria Skłodowska Curie",	"Henryk Sienkiewicz",	"Albert A. Michelson",	"Maria Skłodowska Curie",	"Władysław Reymont",	"Isidor Isaac Rabi",	"Tadeusz Reichstein",	"Andrzej Schally",
       "Menachem Begin",	"Isaac Bashevis Singer",	"Czesław Miłosz",	"Roald Hoffmann",	"Lech Wałęsa",	"Jerzy Charpak",	"Shimon Peres",	"Józef Rotblat",	"Wisława Szymborska",	"Leonid Hurwicz",	"Olga Tokarczuk"]
disc = ["Physics",	"Literature",	"Physics","Chemistry", "Literature",	"Physics", "Medicine",	"Medicine",	"Peace",	"Literature",	"Literature",	"Chemistry",	"Peace",	"Physics",
        "Peace",	"Peace",	"Literature",	"Economics",	"Literature"]

pol_Nobel = list(zip(year, who, disc))
```

```python
import pandas as pd
Nobel_dict = {
  "Rok": year,
  "Laureat": who,
  "Dziedzina": disc
}
df = pd.DataFrame(Nobel_dict)

print("Polscy laureaci pokojowej nagrody Nobla:")
peaceNobel = df[df["Dziedzina"]== "Peace"]
print(peaceNobel)

print()
print("Polscy laureaci literackiej nagrody Nobla po 1950:")
contemporaryLiteratureNobel = df[(df.Dziedzina == "Literature") & (df.Rok > 1980)]
print(contemporaryLiteratureNobel)
```
