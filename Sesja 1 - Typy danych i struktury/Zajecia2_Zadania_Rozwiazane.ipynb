### Zadanie 3

Usuń z listy klubów Premier League drużyny z Liverpoolu (Liverpool, Everton).

Wydrukuj elementy listy oddzielone dwoma gwiazdkami.

**Podpowiedź:** Waato użyć komendy print wraz z rozpakowaniem listy

```python
PremierLeagueTable = ['Chelsea', 'Liverpool', 'Manchester City', 'Manchester United', 'Everton', 'Brighton & Hove Albion', 'Brentford', 'Tottenham Hotspur', 'West Ham United', 'Aston Villa']

# Krok 1: Do usuwania określonych wartości z listy można wykorzystać metodę remove. Wtedy podajemy informację, którą mamy usunąć
PremierLeagueTable.remove('Liverpool')
PremierLeagueTable.remove('Everton')

# Wydrukujmy kształt listy:
print("Lista po usunięciu:", PremierLeagueTable)
```

Lista Kluby_ropa powinna zawierać elementy 'Chelsea' i 'Manchester City'

Lista Kluby_tradycja powinna zawierać elementy 'Manchester United' i 'Tottenham Hotspur'

### Zadanie 4

Popraw kod ISO Włoch (ITA) wykorzystując komendy poznane na zajęciach o typie String.

Użyk kombinacji koment sortowania i odwracania listy, tak aby Polska (POL) znajdowała się na pierwszym miejscu.

```python
lista_skroty = ["KOR", "POL", "GER", "FRA", "ita"]

# Krok 1: Modyfikujemy treść ostatniego elementu za pomocą komendy upper.
lista_skroty[4] = lista_skroty[4].upper()
print("Lista po modyfikacji:", lista_skroty)

# Krok 2: Sortujemy listę - zobaczy, że polska jest na końcu:
lista_skroty.sort()
print("Lista po posortowaniu:", lista_skroty)

# Krok 3: Chcemy odrwrócić kolejność elementów - używamy metody reverse
lista_skroty.reverse()
print("Finalna lista:", lista_skroty, "\n \n")

# Uwaga: Krok 3 realnie jest zbędny - moożemy modyfikować właściwości metody sort
lista_skroty.sort(reverse = True)
print("Lista po posortowaniu v2:", lista_skroty)

```

### Zadanie 5

Poniższy kod wygeneruje listę losowych liczb. Długość listy będzie uzależniona od losowania z 5 długości. Będą się one wahać od 20 do 30 elementów:

```python
# Ten Kod nie wymaga waszej ingerencji - po prostu go uruchomcie
import random
rozmiary_list = [20, 23, 25, 27, 30]

dlugosc_listy = random.choice(rozmiary_list)

losowa_lista = []
for i in range(0,dlugosc_listy):
  n = random.randint(1,30)
  losowa_lista.append(n)

print(losowa_lista)
```

Napisz taki kod, aby zawsze poprawnie drukował:

* 3 od końca element listy
* 5 i 6 od końca element listy

***Podpowiedzi:*** Użyj funkcji len() i ujemnego indeksowania

```python
# Rozwiązanie z funkcją len() - bardzo skomplikowane
print("Trzeci element od końća: ", losowa_lista[len(losowa_lista)-3])
print("Zakres 5 i 6 element od końća: ", losowa_lista[(len(losowa_lista)-6):(len(losowa_lista)-4)])
```

```python
# Rozwiązanie z ujemnym indeksowaniem - dużo lepsze
print("Trzeci element od końća: ", losowa_lista[-3])
print("Zakres 5 i 6 element od końća: ", losowa_lista[-6:-4])
```
