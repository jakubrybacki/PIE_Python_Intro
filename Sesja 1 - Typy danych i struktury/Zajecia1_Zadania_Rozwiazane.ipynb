###Zadanie 1
Wstaw takie wartości a, b i c, aby wszystkie zmienne binarne zwróciły wartość True

Podpowiedzi (równanie 2):
* Instrukcja not zmienia wartość z True na False i odwrotnie.
* Aby zmienna binarna przyjęła wartość False musi odpowiadać jej równanie dające wynik 0.

```python
# Wstawione poprawne odpowiedzi:
a = 12
b = 15
c = 81

# Pierwsze rozwiązanie jest stosunkowo proste - wystarczy dzielić kolejno liczby
#  180 / 5 = 36, następnie 36 / 3 = 12
IsEquation1_Solved =  a == 180/5/3

# Drugie rozwiazanie wymaga wiedzy czym jest operator not
#  not - zmienia wartość True na False i odwrotnie. Potrzebujemy zatem wyniku ktory daje fałsz
#  Aby go uzyskać (b / 3) - 5 powinno wynieść zero. Stad b = 15
IsEquation2_Solved = not((b / 3) - 5)

#Nieco trudniejsze:
# W tym przykładzie w pierwszej kolejności wykonywane jest potęgowanie (dwie gwiazdki)
# 3^2 = 9 - nasze równanie to 9 * 9 = 81
# Oczywiście nigdy w życiu tak tego nie kodujcie.
IsEquation3_Solved =  c == 3**2*3**2

print(IsEquation1_Solved, IsEquation2_Solved, IsEquation3_Solved, sep = '\t')
```

### Zadanie 2
Wykonaj polecenie o które prosi zmienna Text

Podpowiedzi:
* Instrukcja len() pomoże Ci określić ile znaków pobrać.
* Potrzebny będzie operator podwójnego dzielenia (//) oraz odwołanie do konkretnych znaków ([])

```python
Text = "Wydrukuj pierwszych 10% znaków z zawartosci tego tekstu"

# Krok 1.: Ustalmy ilość elementów w tekście
iloscElementow = len(Text)

# Krok 2.: Chcemy wiedzieć ile to 10% znaków - wystarczy podzielić liczbę znaków przez 10
#   Aby uzyskać liczbę całkowitą używamy podwójnego dzielenia
iloscElementow = iloscElementow//10

# Krok 3.: Wybieramy interesujace nas znaki - od pierwszego (z indeksem 0) do naszej ilości
#   elementów
odpowiedz = Text[0:iloscElementow]

# Krok 4.: Teraz wystarczy wydrukować tekst
print(odpowiedz)
```

### Zadanie 4
Zmień rozszerzenie zmiennej fileName z txt na pdf. Usuń też zbędne spacje.

Skonstruuj pełną ścieżkę łącząc zmienne fileDirectory i fileName

```python
filePath = "C:/"

fileName = "  TajnyPlik.txt  "

# Krok 1.: Usuwamy spacje za pomocą komendy strip()
#   Pamiętamy, że wynik takiej operacji należy zapisać do zmiennej.
#   Może to być ta sama zmienna na której operujemy
fileName = fileName.strip()
print("Nazwa pliku po usunięciu spacji:", fileName)

# Krok 2: Zmieniamy rozszerzenie - tutaj posłużymy się komendą replace()
fileName = fileName.replace(".txt", ".pdf")
print("Nazwa po zmianie rozszerzenia: ", fileName)

# Krok 3: Aby złączyć nazwę ścieżki i pliku wystarczy proste dodawanie
odpowiedz = filePath + fileName
print() # Pusta linia
print("Wynik: ", odpowiedz)
```

### Zadanie 5
Rozdziel tekst wykorzystując metodę split (separator |). Wydrukuj kolejne części w instrukcji print. Każde słowo oddziel nową linią.

```python
textToSplit = "Polski | Angielski | Niemiecki"

# Krok 1.: Rozdzielamy tekst za pomocą komendy split()
#   Jej wynikiem będzie lista słow.
lista = textToSplit.split(sep=" | ", maxsplit=-1)

# Krok 2.: Możemy wydrukować każdy element z osoba
#   odwołując się do poszczególnych indeksów.
print("Rozwiązanie brutalne:")
print(lista[0])
print(lista[1])
print(lista[2])
print()

# Krok 3: Są jednak prostsze metody.
#   Aby wydrukować elementy listy wystarczy wpisać * i nazwę listy
#   Pamietamy jednak, że w takim wypadku trzeba ustalić separator
#   Znak enter (nowa linia) opisywany jest przez słowo \n
print("Rozwiązanie wyrafinowane:")
print(*lista, sep = "\n")
```
