## Typy zmiennych oraz operacje - Liczby (int, float)

Celem programowania jest automatyzacja powtarzalnych i pracochłonnych zadań. Działanie wymaga przetwarzania informacji przy wykorzystaniu zmiennych. Ta część będzie mocno teoretyczna, jednak konieczne jest przyswojenie sobie tych podstaw.

Będziemy wykorzystywać:

*   *Integer* - Liczby całkowite
*   *Float* - Ułamki
*   *String* - Tekstowe
*   *Boolean* - Zmienne logiczne (PRAWDA / FAŁSZ)

Typ zmiennych można sprawdzić poleceniem **type**

```python
int_var = 5
float_var = 5.0
str_var = "5"
bool_var = True

print(int_var, float_var, str_var, sep=", ")

print(type(int_var), type(float_var), type(str_var), sep=", ")
```

### Operacje arytmetyczne

Zmienne liczbowe (int, float) pozwalają na prowadzenie operacji arytmetycznych. 
 


```python
print(
    5 + 2, #Dodawanie
    5 - 2, #Odejmowanie
    5 * 2, #Mnożenie
    5 / 2, #Dzielenie
    5 % 2, #Modulo, czyli reszta z dzielenia
    5 ** 2, #Potęga
    sep = "\n" #separator \n wydrukuje każdy wynik w nowej linii
    )
```

Przy wykonywanych operacjach obowiązuje kolejność wykonywania działań.

* Równorzędne operacje (np. mnożenie i dzielenie) wykonywane są od lewej do prawej strony.
* Wyjątek: potęgowanie wykonywane jest przed mnożeniem / dzieleniem.

```python
print(
    "Przykład - dzielenie:",
    10 / 5 / 2,
    10 / (5/2),
    
    "\nPrzykład - potęgowanie:",
    2*5**3,
    (2*5)**3,
    sep = "\n")
```

Przykład zastosowania operacji arytmetycznych bezpośrednio  na zmiennych:

```python
int_var = int_var // 2
float_var = float_var / 2
print(int_var, float_var, sep="\t ") #separator \t to odpowiednik wciśnięcia klawisza tab
```

*Ciekawostka:* Podwójny znak dzielenia podpowiada, kompilatorowi że chcemy zachować liczbę całkowitą. W przypadku dzielenia maszyna automatycznie zmieni typ liczby na zmiennoprzecinkową.

```python
int_var_converted = 5/2
int_var_notConverted = 5 // 2 
print(type(int_var_converted), type(int_var_notConverted))
```

### Operacje logiczne

Chociaż programowanie i Python można wykorzystywać jako uniwersalny kalkulator, będzie to jednak duże zmarnowanie jego potencjału. Wyniki obliczeń zazwyczaj wykorzystywać będziemy do podejmowania decyzji co robić dalej - tutaj przydatne są operacje logiczne. 

Ich wynik będzie jedną z dwóch wartości:
* Prawda (True)
* Fałsz (False)

```python
print(
    5 > 2, 
    5 >= 2, 
    5 < 2, 
    5 <= 2, 

    5 == 2 + 3, #Identyczność
    5 != 2 + 3, #Różny od
    
    sep = "\n" #separator \n wydrukuje każdy wynik w nowej linii
    )
```

Proste zadanie - potrzebujemy informacji czy z 3 odcinków a,b i c o różnych długościach możemy złożyć trójkąt.  

**Szybki Quiz:** Sekcja Booleans z [W3 Schools](https://www.w3schools.com/python/exercise.asp?filename=exercise_booleans1)

```python
a = 3
b = 9
c = 5

#Warunek istnienia trójkątu - każdy odcinek jest krótszy od sumy dwóch pozostałych
firstLineExist = a + b > c
secondLineExist = a + c > b
thirdLineExist = b + c > a

print(firstLineExist, secondLineExist, thirdLineExist, sep="\t")

#Ujmijmy to w jeden warunek
isTriangle = firstLineExist and secondLineExist and thirdLineExist

print(isTriangle)
```

##Typ string i metody

Ta sekcja przybliży operacje jakie można wykonywać na zmiennych tekstowych. Najczęsciej wykorzystywaną operacją jest zapytanie o rozmiar tekstu. Do jego określenia używamy funkcji ***len()*** od angielskiego length.

```python
text = "Zwykly tekst"
len(text)
```

Często interesować będą nas określone partie tekstu - do ich wyciągnięcia wykorzystujemy operator [].

***Uwaga***: Python rozpoczyna numerowanie znaków od 0. 

```python
text[0:6]
```

Zmienne typu string posiadają też zestaw własnych funkcji zwanych metodami. Aby je wykorzystać należy zapisać nazwę danego obiektu, kropkę oraz nazwę funkcji. 

Przykład: Zamawiacie badanie ankietowe gdzie, osoby deklarują np. wyznawaną wiarę. Osoby, które przepisywały wyniki z formularzy do bazy danych zrobiły to niechlujnie - pojawiają się częste niepotrzebne spacje. 

Do rozwiązania tego wykorzystamy funkcje ***strip()***. Usuwa ona nadmiarowe spacje, tabulatory z początku i końca tekstu. 

```python
textWithSpaces = "    Złoto     "
cleanText = textWithSpaces.strip()
print(cleanText)

#Dla sprawdzenia - czy rzeczywiscie spacje zostały usunięte 
print(len(textWithSpaces), len(cleanText))
```

Przykład 2: Prowadzisz serwis w którym udostępniasz dane albo publikacje dla klientów zwenętrznych. Opiekunowie klienta rejestrują loging jako mail. Robią to jednak dość automatcznie. Niektóre maile zaczynają się z dużej litery, niektóre z małej. Klient nie wie jakiej litery użył opiekun - po prostu wpisze swój adres.

Jak wyjść z sytuacji? Zazwyczaj systemy bazodanowe z automatu przekształcają adres e-mail na małe bądź wielkie litery. Służą do tego metody ***lower()*** i ***upper()***

```python
mailAddress = "Jakub.Rybacki@uw.edu.pl"
print(mailAddress.lower(), mailAddress.upper()) 
```

### Formatowanie tekstu

Przykład 3: Prowadzimy cykliczny raport gdzie opisujemy zmiany cen surowców. Jak uniknąć żmudnego przepisywania podobnych tekstów? 

Przyda się nam metoda ***format()***. Pozwala ona wstawic w miejsca określone przez {} dowolne wartości.

```python
textToPrint = "{} drożeje na rynku światowym. Ceny wzrosły o {}% w trakcie dzisiejszej sesji."
print(textToPrint.format("Ropa naftowa", str(5)))
print(textToPrint.format("Złoto", str(1)))
```

Wskazówka: Metoda format doczekała się specjalnego słowa kluczowego, aby nie trzeba było każdorazowo deklarować zmiennej. 

```python
CL_Winner = "Chelsea Londyn"
Opponent = "Juventus"
print(f"{CL_Winner} to pretendent w walce o Ligę Mistrzów. {Opponent} czeka ciężkie zadanie.")
```

### Praca z tekstami - metody split, replace

Ta część będzie zadaniowa. Pokaże jak za pomocą kilku komend możemy oszczędzić sobie dużych wysiłków przy przetwarzaniu danych.


Naszym celem jest przygotowanie bazy danych o zmianach cen materiałów budowlanych na podstawie komunikatów prasowych [Grupy PSB](https://www.grupapsb.com.pl/centrum-prasowe/trendy-cenowe). 


Dane umieszczane są w formie tekstowej, kolejność kategorii pojawiających się w tekście jest różna, chociaż trzyma się pewnych reguł. Przepisywanie tego ręcznie możę doprowadzić do nerwicy. Tu wchodzi programowanie. 

Przykład zawiera jeden tekst:

```python
PSBText = "Ceny w lipcu 2021 r., w stosunku do lipca 2020 r., średnio wzrosły o 8,3%. Odnotowano wzrosty we wszystkich 20 grupach towarowych: płyty OSB (+27,5%), izolacje termiczne (+24,1%), dachy, rynny (+12,5%), instalacje, ogrzewanie (+10,9%), otoczenie domu (+10,4%), narzędzia (+9,1%), wykończenia (+9,0%), ogród i hobby (+8,8%), płytki, łazienki, kuchnie (+8,2%), izolacje wodochronne (+7,9%), oświetlenie, elektryka (+7,7%), wyposażenie, AGD (+6,8%), sucha zabudowa (+6,8%), motoryzacja (+5,8%), chemia budowlana (+5,5%), dekoracje (+5,2%), farby, lakiery (+4,9%), stolarka (+4,6%), cement, wapno (+3,8%) oraz ściany, kominy (+2,1%)."

```

Obserwacja 1: Liczby opisujące zmiany cen w poszczególnych kategoriach produktów pojawiają się po dwukropku - konieczne jest rozdzielenie tych tekstów. Wykorzystujemy do tego metodę ***split()***. 

```python
PSBSplitText = PSBText.split(sep=":", maxsplit=1)
print(PSBSplitText[0], PSBSplitText[1], sep = "\n")
```

Dalej pracujemy z drugim członem. Kategorie porozdzielane są przecinkami, ale nie możemy użyć tego jako separatora - niepoprawnie rozbijemy miejsca po przecinku i część nazw kategori. Zaaplikumjemy inny trik - każdy nawias poza ostatnim kończy się sekwencją "%),". 

Najpierw dostosujmy ostatnią część korzystając z metody ***replace()***

```python
PSBSplitText[1] =  PSBSplitText[1].replace(" oraz", ",")
PSBSplitText[1] =  PSBSplitText[1].replace(".", ",")
print(PSBSplitText[1])
```

Robimy kolejny split - w efekcie dostaniemy pełną listę produktów. Operacje na listach omówimy na kolejnych zajęciach - teraz mała próbka:

```python
PSBSplitCategories = PSBSplitText[1].split(sep="%),", maxsplit=-1)
PSBSplitCategories.pop()
[print(i) for i in PSBSplitCategories]

type(PSBSplitCategories)
```

Pozostaje ostatni szlif - liczby zaczynają się od nawiasu - kolejny split. Wykonanie takiego splitu na wszystkich elementach listy jest prostsze przy wykorzystaniu biblioteki pandas.

```python
import pandas
PSBSplitCategoriesPD = pandas.DataFrame(PSBSplitCategories)

PSBSplitCategoriesPD[["Nazwa", "Wartosc"]] = PSBSplitCategoriesPD[0].str.split("(", 1, expand=True)

display(PSBSplitCategoriesPD)
```

Pozostaje usunąć: 

*   kolumne z tekstem źródłowy
*   nadmiarowe puste znaki w kolumnie nazwa

oraz zmienić przecinek na kropkę w kolumnie wartość

```python
PSBSplitCategoriesPD = PSBSplitCategoriesPD[["Nazwa", "Wartosc"]]
PSBSplitCategoriesPD["Nazwa"] = PSBSplitCategoriesPD["Nazwa"].str.strip()
PSBSplitCategoriesPD["Wartosc"] = PSBSplitCategoriesPD["Wartosc"].str.replace(",",".")

display(PSBSplitCategoriesPD)
```
