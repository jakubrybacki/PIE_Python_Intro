Oraz inicjujemy klienta Youtube:

# Zadanie 1: Komentarze
Wyszukaj pierwsze 50 komentarzy do filmu z reportażu radio ZET
1. Przyda się karta dotycząca wątków w komentarzach - [link](https://developers.google.com/youtube/v3/docs/commentThreads/list)
2. Zajmiemy się filmem o ID: *_9UXV1qniAk* - Gość Radio ZET

```python
request = youtube.commentThreads().list(
    part="snippet",
    videoId="_9UXV1qniAk"
)
response = request.execute()

print(response)
```

Przekształcamy dane do ramki:

```python
dane = response["items"] # WYBIERZ DANE Z POZYCJI ITEMS 
response_data_df = pd.DataFrame(dane) # PRZEKSZTAŁC TO W FORMAT ARKUSZA I ZAPISZ DO ZMIENNEJ response_data_df
response_data_df
```

Zobaczymy kolejne elementy ramki: 

```python
df2 = pd.json_normalize(response_data_df['id']) # PRZEKSZTAŁC CZĘŚĆ ID DO CZYTELNYCH DANYCH
df3 = pd.json_normalize(response_data_df['snippet']) # PRZEKSZTAŁC CZĘŚĆ SNIPPET DO CZYTELNYCH DANYCH
wynik =  pd.concat([df2, df3], axis=1) # ZLACZ CZESCI ID i SNIPPET W JEDNA RAMKE
wynik
```

Wydrukujemy sobie tylko nazwisko komentującego oraz tekst komentarza: 

```python
wynik[["topLevelComment.snippet.authorDisplayName",
      "topLevelComment.snippet.textDisplay"]]	
```

# Zadanie 2: Playlisty Radia ZET
Wyszukajmy wszystkie playlisty Radio ZET, następnie przeszukajmy filmy z sekcji Gość Radio ZET. 
* Konto Radio ZET ma ID: *UCvHFbkohgX29NhaUtmkzLmg*
* Dokumentacja playlisty jest dostępna pod następującym [linkiem](https://developers.google.com/youtube/v3/docs/playlists/list)

```python
request = youtube.playlists().list(
    part="snippet",
    channelId="UCvHFbkohgX29NhaUtmkzLmg",
    maxResults = 50
)
response = request.execute()

pd.json_normalize(response)
```

```python
dane = response["items"] # WYBIERZ DANE Z POZYCJI ITEMS 
response_data_df = pd.DataFrame(dane) # PRZEKSZTAŁC TO W FORMAT ARKUSZA I ZAPISZ DO ZMIENNEJ response_data_df
response_data_df
```

```python
df2 = pd.json_normalize(response_data_df['id']) # PRZEKSZTAŁC CZĘŚĆ ID DO CZYTELNYCH DANYCH
df3 = pd.json_normalize(response_data_df['snippet']) # PRZEKSZTAŁC CZĘŚĆ SNIPPET DO CZYTELNYCH DANYCH
wynik =  pd.concat([response_data_df['id'], df2, df3], axis=1) # ZLACZ CZESCI ID i SNIPPET W JEDNA RAMKE
wynik
```
