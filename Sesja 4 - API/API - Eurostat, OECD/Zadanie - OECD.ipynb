# Zadanie - Pobieranie danych z OECD

W tym zadaniu wykorzystamy funkcję łączące oraz paczkę pobierającą dane z OECD. Jest ona udostępniana przez osoby trzecie za pośrednictwem [Github](https://github.com/expersso/OECD)

Jej instalację uruchomi nastepująca komenda:

```r
remotes::install_github("https://github.com/expersso/OECD")
```

Następnie załączamy bibliotekę tidyverse oraz paczkę OECD:

```r
require("OECD")
require("dplyr")
```

Sprawdź dostępne zbiory danych - wykorzystaj pomoc paczki OECD oraz strony z [eksploratorem danych](https://data-explorer.oecd.org/). Sama strona dokumentacji API to ten [link](https://data.oecd.org/api/sdmx-json-documentation/)

```r
help(p = "OECD")
```

Znajdz nazwy serii zawierających dane z publikacji Economic Outlook.

Wykorzystajmy metodę *get_data_structure*, aby znaleźć kod serii opisujący *output gap* - To niestety nie zawsze działa w paczce, ale można samemu znaleźć na stronie API [OECD](https://stats.oecd.org/restsdmx/sdmx.ashx/GetDataStructure/EO112_INTERNET/)



```r
?get_dataset
```

```r
get_data_structure("OECD.ECO.MAD,DSD_EO@DF_EO,1.0")
```
