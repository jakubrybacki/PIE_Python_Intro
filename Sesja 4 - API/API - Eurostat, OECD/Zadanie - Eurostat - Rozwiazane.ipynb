### Polecenie 2:
Wykorzystaj funkcje do sciagniecia tabeli hbs_str_t223

Uwaga: Wybrana tabela jest mala - wystarczy argument z tytulem
W przypadku zapytan o inne dane warto budowac filtr z argumentami.

```r
df <- get_eurostat("hbs_str_t223")
head(df)
```

## Polecenie 3:
Sposrod tych danych:
1. wybierz te, ktore dotycza pierwszego kwintyla dochodów oraz wylacznie wyniki dla 2015 roku.
2. Wybierz 3 kolumny: geo, coicop oraz values

```r
colnames(df)
```

```r
dane <- df %>% filter(quantile == "QUINTILE1",
                      TIME_PERIOD == "2015-01-01") %>%
               select(geo, coicop, values)

head(dane)

dim(dane)
```

## Polecenie 4:
Sposrod tych danych wybierz te, gdzie kolumna coicop ma 4 litery:

```r
dane2 <- dane %>% filter(nchar(coicop)== 4)

head(dane2)
dim (dane2)
```

## Polecenie 5:
Przeksztalc te dane z formy dlugiej do krotkiej.

```r
require(tidyr)
dane_dlugie <- dane2 %>% pivot_wider(names_from = "geo", values_from = "values")

dane_dlugie
```
