```python
password = "Tajne"
tekst = input("Wpisz hasło:")
if tekst.find("'") >= 0:
  print("Wyryto atak")
else:
  # Częsty błąd z zajęć - pamiętajcie, że porównanie wymaga dwóch znaków =.
  if tekst == password:
    print("Witaj w systemie")
  else:
    print("Nieprawidłowe hasło")
```

### Zadanie 2
Napisz program, który:

* Rozbije tekst piosenki na poszczególne wyrazy - wykorzystaj metodę *split* z pierwszych zajęć.
* W pętli doda do listy *slowa4plus* tylko te wyrazy, które zawierają przyanjmniej 4 znaki.


```python
tekst_piosenki = "Byłem sam I leżałem przyparty do szyn (Piorun) Pociąg gnał I nie moglem już nic zrobić z tym"
lista_slow = tekst_piosenki.split(sep=" ", maxsplit=-1)
slowa4plus = []

# Podejście 1 - w pętli przeglądamy listę. Dodajemy element do listy, jeżeli spełniony jest warunek dotyczacy długości.
for element in lista_slow:
  if len(element) > 3:
    slowa4plus.append(element)

#print(*slowa4plus, sep = "\n")

# Podejście 2 - wykorzystujemy szybką pętlę z warunkiem dotyczącym długości słowa
slowa4plus2 = [element for element in lista_slow if len(element) > 3]
print(*slowa4plus2, sep = "\n")
```

### Zadanie 3
Wykorzystaj szybką pętlę, aby z listy liczb wybrać tylko te podzielne przez 3.


```python
liczby = [1, 3, 5, 7, 9, 11, 14, 15, 19, 21, 25, 27, 39]

# Resztę z dzielenia zwraca operator modulo oznaczany przez %. Podzielność przez 3 oznacza brak reszty.
wybrane_liczby = [element for element in liczby if element % 3 == 0]
print(wybrane_liczby)
```

### Zadanie 5
Napisz program który będzie analizować listę cyfr w pętli. Potrzebne są następujące instrukcje (Program):

* Przejdzie do następnego wykonania pętli jeżeli liczba jest większa od 10 - użyj instrukcji *continue*
* Przerwie pętle w momencie kiedy liczba jest większa od 100 - użyj instrukcji *break*
* doda do wyników trzecią potęge liczby z listy oraz wyświetli analizowaną wartość

**Rozwiązanie:**

Zacznijmy od końca - zróbmy na razie listę która podniesie każdy element do trzeciej potęgi. Kod wyglądać będzie następująco:

```python
ListaLiczb = [5, 12, 4, 22, 9, 111, 1, 3 ,6]
wyniki = []

for element in ListaLiczb:
  print(element ** 3)
  wyniki.append(element ** 3)
```

Teraz odfiltrujmy wielkie liczby - zróbmy warunek, który porzuci wykonywanie obliczeń jeżeli liczba jest większa od 10.

```python
ListaLiczb = [5, 12, 4, 22, 9, 111, 1, 3 ,6]
wyniki = []

for element in ListaLiczb:
  if element > 10:
    continue

  print(element ** 3)
  wyniki.append(element ** 3)
```

Na koniec chcemy przerwać wykonywanie pętli w momencie, kiedy pierwszy element przekracza 100.

Zwróćmy uwagę, że konieczne jest zachowanie kolejności w sprawdzanych warunkach.

```python
ListaLiczb = [5, 12, 4, 22, 9, 111, 1, 3 ,6]
wyniki = []

for element in ListaLiczb:
  if element > 100: # UWAGA: Jeżeli zmienilibyśmy kolejność warunków, ten kod nigdy nie zostałby wykonany
    break

  if element > 10:
    continue

  print(element ** 3)
  wyniki.append(element ** 3)

print("\n")
print("Finalna lista:")

print(*wyniki, sep = "\n")
```
